## A demonstration on an approach to data-orchestration



```python
import httpx
```

```python
import os
```

From the [docs](https://developer.ticketmaster.com/products-and-docs/apis/discovery-api/v2/#search-events-v2) 

* The default quota is 5000 API calls per day and rate limitation of 5 requests per second.
* Deep Paging: we only support retrieving the 1000th item. i.e. ( size * page < 1000)




```python
TM_API_KEY = os.environ['TM_API_KEY']
```

```python
latlong = '40.7659051,-111.8487420'  # SLCPython UoU
radius = 60
unit = 'miles'
page = 1
locale = "*"
```

```python
url = f"https://app.ticketmaster.com/discovery/v2/venues?apikey={TM_API_KEY}&latlong={latlong}&radius={radius}&unit={unit}&locale={locale}&page={page}"
```

```python
response = httpx.get(url)
```

```python
response
```

```python
response.json()
```

```python
import pandas as pd

df = pd.DataFrame.from_records(response.json()['_embedded']['venues'])
df
```

```python
start_dt = "2024-03-06T15:04:00Z"
url = f"https://app.ticketmaster.com/discovery/v2/events?apikey={TM_API_KEY}&latlong={latlong}&radius={radius}&unit={unit}&locale={locale}&startDateTime={start_dt}"
url
```

```python
resp = httpx.get(url)
```

```python
resp
```

```python
resp.json()
```

```python
df = pd.DataFrame.from_records(resp.json()['_embedded']['events'])
df
```
